```python
import urllib.request, json
import re
import folium
from folium.plugins import MarkerCluster
```

```python
## Read json
regions = 'https://raw.githubusercontent.com/dimitris1ps/Geojson/master/oneToThreeGreeceRegion.geojson'

with urllib.request.urlopen(re.sub('\\n|\s+', '', regions)) as url:
    greece_regions_json = json.loads(url.read().decode())
```

```python
regions_mapping = {'Νότιο Αιγαίο':'#543c00', 
                   'Δυτική Μακεδονία':'#a57f1f',
                   'Ήπειρος':'#375e30',
                   'Θεσσαλία': '#157f00',
                   'Aττική':'#8c0416',
                   'Βόρειο Αιγαίο':'#4286f4',
                   'Ιόνια Νησιά':'#005454',
                   'Δυτική Ελλάδα':'#903f96',
                   'Στερεά Ελλάδα':'#c4e5a9',
                   'Πελοπόννησος':'#493b3d',
                   'Κρήτη':'#3f9696',
                   'Aνατολική Μακεδονία, Θράκη':'#9ed1e5',
                   'Κεντρική Μακεδονία':'#68476b'}
```

```python
def fillcolour(feature):
    """Maps regions to colours"""
    # some criteria here to pick a colour
    # This will return the region name (regions_mapping[feature['properties']['name']]) e.g κρήτη
    return regions_mapping[feature['properties']['name']]
```

```python
tiles = ['cartodbpositron', 'OpenStreetMap', 'Stamen Watercolor', 'mapquestopen', 'MapQuest Open Aerial', 
        'Mapbox Bright', 'Mapbox Control Room', 'stamenterrain', 'stamentoner', 'stamenwatercolor', 
        'cartodbdark_matter']
```

```python
m = folium.Map(width='100%',height=800,location=[38,25], zoom_start=7,min_zoom = 6, tiles=tiles[9])
```

```python
folium.GeoJson(greece_regions_json, 
               tooltip=folium.GeoJsonTooltip(fields=['name'],aliases=['Region']), 
               style_function=lambda feature:{'fillColor': fillcolour(feature), 'color': 'black', 'weight': 0.8, 'fillOpacity':0.6}
              ).add_to(m);
```

```python
folium.plugins.Fullscreen(position='topright',title='Expand me', title_cancel='Exit me',
                          force_separate_button=True).add_to(m);
```
