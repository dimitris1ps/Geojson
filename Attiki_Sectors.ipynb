```python
import urllib.request, json
import re
import folium
from folium.plugins import MarkerCluster
```

```python
## Read json
regions = 'https://raw.githubusercontent.com/dimitris1ps/Geojson/master/attiki.geojson'

with urllib.request.urlopen(re.sub('\\n|\s+', '', regions)) as url:
    greece_regions_json = json.loads(url.read().decode())
```

```python
regions_mapping = {'Πειραιάς, Νήσοι':'#543c00', 
                   'Νότιος Τομέας Αθηνών':'#a57f1f',
                   'Ανατολική Αττική':'#375e30',
                   'Δυτική Αττική': '#157f00',
                   'Βόρειος Τομέας Αθηνών':'#8c0416',
                   'Δυτικός Τομέας Αθηνών':'#4286f4',
                   'Κεντρικός Τομέας Αθηνών':'#005454'}
```

```python
def fillcolour(feature):
    """Maps regions to colours"""
    # some criteria here to pick a colour
    # This will return the region colour (regions_mapping[feature['properties']['name']]) e.g κρήτη
    return regions_mapping[feature['properties']['NUTS_NAME']]
```

```python
tiles = ['cartodbpositron', 'OpenStreetMap', 'Stamen Watercolor', 'mapquestopen', 'MapQuest Open Aerial', 
        'Mapbox Bright', 'Mapbox Control Room', 'stamenterrain', 'stamentoner', 'stamenwatercolor', 
        'cartodbdark_matter']
```

```python
m = folium.Map(width='100%',height=800,location=[38,23.75], zoom_start=10,min_zoom = 6, tiles=tiles[1])
```

```python
folium.GeoJson(greece_regions_json, 
               tooltip=folium.GeoJsonTooltip(fields=['NUTS_NAME'],aliases=['Sector']), 
               style_function=lambda feature:{'fillColor': fillcolour(feature), 'color': 'black', 'weight': 0.8, 'fillOpacity':0.6}
              ).add_to(m);
```

```python
folium.plugins.Fullscreen(position='topright',title='Expand me', title_cancel='Exit me',
                          force_separate_button=True).add_to(m);
```
